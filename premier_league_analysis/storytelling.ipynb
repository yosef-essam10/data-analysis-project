```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fpl_playerstats_2024-25.csv')
df
```

```python
df.columns
```

```python
print("min point:", df['total_points'].min())
print("max point:", df['total_points'].max())
```

```python
df['value_for_money'] = df['total_points'] / df['player_cost']
```

```python
df
```

```python
mean = df['total_points'].mean()
median = df['total_points'].median()
mode = df['total_points'].mode().iloc[0]
```

```python
sns.histplot(df['total_points'], bins=30, kde=True)
plt.title("Distribution of Total Points (2024–2025)")
plt.xlabel("Total Points")
plt.ylabel("Number of Players")
plt.savefig("1_total_points_distribution.png")
plt.show()
```

```python
sns.histplot(df['total_points'], bins=30, kde=True)
plt.axvline(mean, color='red', linestyle='--', label=f"Mean: {mean:.1f}")
plt.axvline(median, color='green', linestyle='--', label=f"Median: {median}")
plt.axvline(mode, color='blue', linestyle='--', label=f"Mode: {mode}")
plt.title("Distribution of Total Points with Mean, Median, Mode")
plt.xlabel("Total Points")
plt.ylabel("Number of Players")
plt.legend()
plt.savefig("1_total_points_with_avg.png")
plt.show()
```

```python
counts, bins = np.histogram(df['total_points'], bins=30)
for i in range(len(counts)):
    print(f"Bin {i+1}: from {bins[i]:.1f} to {bins[i+1]:.1f} --> {counts[i]} player")
```

```python
top10 = df.nlargest(10, 'value_for_money')
sns.barplot(x='value_for_money', y='second_name', data=top10)
plt.title("Top 10 Value-for-Money Players")
plt.xlabel("Value (Points per Cost)")
plt.ylabel("Player")
plt.savefig("2_top_value_for_money.png")
plt.show()
```

```python
sns.scatterplot(x='minutes', y='total_points', hue='player_position', data=df)
plt.title("Minutes Played vs Total Points")
plt.xlabel("Minutes Played")
plt.ylabel("Total Points")
plt.savefig("3_minutes_vs_points.png")
plt.show()
```

```python
sns.scatterplot(x='player_form', y='total_points', hue='player_position', data=df)
plt.title("Player Form vs Total Points")
plt.xlabel("Current Form")
plt.ylabel("Total Points")
plt.savefig("4_form_vs_points.png")
plt.show()
```

```python
top_assisters = df.nlargest(10, 'assists')
sns.barplot(x='assists', y='second_name', data=top_assisters)
plt.title("Top 10 Players by Assists")
plt.savefig("5_top_assists.png")
plt.show()
```

```python
top_scorers = df.nlargest(10, 'goals_scored')
sns.barplot(x='goals_scored', y='second_name', data=top_scorers)
plt.title("Top 10 Goal Scorers")
plt.savefig("6_top_scorers.png")
plt.show()
```

```python
sns.scatterplot(x='expected_goals', y='goals_scored', hue='player_position', data=df)
plt.title("Expected Goals vs Actual Goals")
plt.xlabel("Expected Goals (xG)")
plt.ylabel("Goals Scored")
plt.savefig("7_expected_vs_actual_goals.png")
plt.show()
```

```python
gw_cols = [col for col in df.columns if col.startswith('gw')]
df['gw_avg'] = df[gw_cols].mean(axis=1)
df['predicted_next_gw'] = df['gw_avg']

top_predicted = df[['first_name', 'second_name', 'predicted_next_gw']].sort_values(
    by='predicted_next_gw', ascending=False).head(10)

print("\n Top 10 Players - Expected Points in Next Gameweek:")
print(top_predicted.to_string(index=False))
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='predicted_next_gw', y='second_name', data=top_predicted)
plt.title('Top 10 Players - Predicted Next GW Points')
plt.xlabel('Expected Points')
plt.ylabel('Player')
plt.savefig("8_Top 10 Players - Predicted Next GW Points.png")
plt.show()
```

```python
team_points = df.groupby('team_name')['total_points'].sum().sort_values(ascending=False).reset_index()

print("\n Total Points by Team:")
print(team_points.to_string(index=False))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='total_points', y='team_name', data=team_points)
plt.title('Total Points by Team')
plt.xlabel('Total Points')
plt.ylabel('Team')
plt.savefig("9_Total Points by Team.png")
plt.show()
```

```python
discipline = df[['second_name', 'yellow_cards', 'red_cards']].sort_values(
    by=['yellow_cards', 'red_cards'], ascending=False).head(10)

print("\n Top 10 Players with Most Yellow/Red Cards:")
print(discipline.to_string(index=False))
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='yellow_cards', y='second_name', data=discipline)
plt.title('Top 10 Players - Yellow Cards')
plt.xlabel('Yellow Cards')
plt.ylabel('Player')
plt.savefig("10_Top 10 Players - Yellow Cards.png")
plt.show()
```

```python
team_points = df.groupby("team_name")["total_points"].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(
    team_points,
    labels=team_points.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Team Points Distribution (FPL 2024–2025)")
plt.axis("equal")  
plt.tight_layout()
plt.savefig("11_Team Points Distribution (FPL 2024–2025).png")
plt.show()
```
